```python
class restrictedType(type):
    def __setattr__(self, attr, value):
        if hasattr(self, "rawCopy") and self.rawCopy == True:
            super().__setattr__(attr, value)
            del self.rawCopy

        if hasattr(self, f"__initialised") and self.__initialised == True:
            if attr in self.__dict__:
                raise ValueError(
                    f"Cannot Reassign Dynamic Attribute on a non copy customModuleType: attr ({attr})"
                )
        super().__setattr__(attr, value)


patchable = restrictedType("patchable_mod", (ModuleType,), {})
```
